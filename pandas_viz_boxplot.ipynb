```python
import pandas as pd
%matplotlib inline
pd.__version__
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(10)
```

```python
df.cty.plot.box()
```

```python
df.boxplot()
```

```python
df[['cty', 'hwy']].plot.box()
```

```python
df[['cty', 'hwy']].boxplot();
```

```python
df.boxplot(column=['cty', 'hwy'])
```

```python
df.boxplot(column=['cty', 'hwy'], by='drv')
```

```python
df.boxplot(column=['cty', 'hwy'], by='drv', grid=False)
```

```python
df.boxplot(column='cty', by=['class', 'drv'])
```

```python
df.boxplot(column='cty', by=['class', 'drv'], rot=90, grid=False)
```

```python
df.boxplot(column='cty', by=['class', 'drv'], vert=False, grid=False);
```

```python
df.boxplot(column=['cty', 'hwy'], by=['class', 'drv'], vert=False, grid=False);
```

```python
df.boxplot(column=['cty', 'hwy'], by=['class', 'drv'], vert=False, grid=False, figsize=(10, 5));
```
